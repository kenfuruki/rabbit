### pandasでデータフレームとしてみてみる

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df["target"] = iris.target
df.head()
```

```python
df.tail()
```

```python
# seabornで各特徴量ごとの相関をみてみる
sns.pairplot(df, hue="target", palette="winter")  # 引数hueはその指定した列の値ごとに色分けする
```

### 1.可視化 null確認
#### nullの数がどのくらいあるか確認する。irisはデータが整っているのでゼロ。

```python
res = df.describe()
#1 nullの可視化
res[res.index == "count"].append(df.isna().sum(), ignore_index=True).T.plot.bar(stacked=True, legend=False)
```

### 2.分布把握系
#### 2-1 平均と偏差

```python
res[res.index == "mean"].T.plot.bar(yerr=res[res.index=="std"].values, capsize=2)
```

#### 2-2 四分位と最小最大

```python
target_name = ["min", "25%", "50%", "75%", "max"]
res[res.index.isin(target_name)].boxplot(figsize=(10,6))
```

#### 2-3 分布

```python
df[res.columns].plot(subplots=True, kind="kde",figsize=(16,6)) # kdeは確率密度分布
```
