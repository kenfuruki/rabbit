# iris SVM

### 1. ライブラリインポート

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

### 2. データの読み込み

```python
iris = datasets.load_iris()
X = iris.data[:,:2]
Y = iris.target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3,random_state=0)
```

### 3. モデル構築

```python
# model構築
model1 = SVC(kernel="linear")
model2 = SVC(kernel="rbf")

model1.fit(X_train, Y_train)
model2.fit(X_train, Y_train)
```

```python
# 線形カーネルの結果
fig = plt.figure(figsize=(8,5))
plot_decision_regions(X_train, Y_train, clf = model1)
plt.show()
```

```python
# rbfカーネルの結果
fig = plt.figure(figsize=(8,5))
plot_decision_regions(X_train, Y_train, clf = model2)
plt.show()
```

### 4. 評価

```python
from sklearn import metrics
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
pre1 = model1.predict(X_test)
pre2 = model2.predict(X_test)
```

```python
#線形カーネルの結果
fig = plt.figure(figsize=(8,5))
plot_decision_regions(X_test, pre1, clf = model1)
plt.show()
```

```python
print(Y_test)
print(pre1)
```

```python
# 正解率
ac_score1 = metrics.accuracy_score(Y_test, pre1)
ac_score2 = metrics.accuracy_score(Y_test, pre2)
print(f"ac_score1(linear) = {ac_score1:.3f}, ac_score2(rbf) = {ac_score2:.3f}")
```
